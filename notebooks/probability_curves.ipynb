```python
import numpy as np
import matplotlib.pyplot as plt

def probability_B(params, y):
    return 1 / (1 + np.exp(params['a_B'] - params['b_B'] * y))
    
def probability_E(params, y):
    return 1 / (1 + np.exp(params['a_E'] - params['b_E'] * y))
```

```python

params = {
    'b_B': 0.01,
    'a_B': 1.5,
    'b_E': 0.01,
    'a_E': 0.7
}

y_B = np.linspace(0, 150, 100)
y_E = np.linspace(0, 150, 100)

p_B = probability_B(params, y_B)
p_E = probability_E(params, y_E)

plt.figure(figsize=(10, 6))
plt.plot(y_B, p_B, label='Business Class Probability', color='blue')
plt.plot(y_E, p_E, label='Economy Class Probability', color='green')
plt.xlabel('Price ($)')
plt.ylabel('Probability')
plt.title('Probability of Buying a Ticket as a Function of Price')
plt.legend()
plt.show()

```
