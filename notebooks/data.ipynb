```python
import dask.dataframe as dd

# Initialize Dask (optional)
from dask.distributed import Client
client = Client()  # This will start a local cluster

# Define the path to your Parquet file
parquet_file = 'data/itineraries_gzip.parquet'

# Read the Parquet file with Dask
df = dd.read_parquet(parquet_file)


```

```python
# Take the 10 legIds with the most rows
legIds = df['legId'].value_counts().compute().head(10).index.tolist()

# Save each of them to separate CSV files
for legId in legIds:
    df[df['legId'] == legId].compute().to_csv(f'data/legs/{legId}.csv', index=False)

```
